```python
le_map = {}  # keep encoders per column to reuse on test

for col in train_df.select_dtypes(include=["object"]).columns:
    if col == "satisfaction":
        continue
    le = LabelEncoder()
    train_df[col] = le.fit_transform(train_df[col].astype(str))
    if col in test_df.columns:
        test_df[col] = le.transform(test_df[col].astype(str))
    le_map[col] = le
```

The above code is an Encoding Technique.

In the first line, it looks for all columns in the dataset that are of type object (text/categorical).

These columns are then converted into numbers because machine learning models cannot work with text.

The â€œsatisfactionâ€ column (our target: satisfied/unsatisfied) is left out for now, since we will encode it separately at the end.

X_train and y_train

When we prepare data for ML, we split it into features (X) and target (y):

X (features): all the input columns we use to make predictions.
Example in your airline dataset:
Age, Flight Distance, Inflight wifi service, Seat comfort, Cleanliness, â€¦

y (target): the column we want to predict.
Example:
satisfaction (Satisfied / Unsatisfied)
Train vs Test

X_train: part of the features used to train the model.

y_train: part of the target values used to train the model.

X_test: the remaining features used to test the model.

y_test: the remaining target values used to check predictions.
